3.1 Постройте график в абсолютных величинах

```python
df5 = pd.crosstab(index=df['Company'],columns=df['Cpu'], normalize = 'index')
df5 = df5.drop(columns=['Samsung']).reset_index()
df5 = df5.set_index('Company')
df5
```

```python

import numpy as np

plt.figure(figsize=(6,4))
number_ticks = np.arange(len(df5['AMD']))
offset1 = 0.2
col_width = 0.4

plt.bar(number_ticks - offset1, df5['AMD'], width= col_width)
plt.bar(number_ticks + offset1, df5['Intel'], width= col_width)
plt.title('Взаимосвязь')
plt.xlabel('Company')
plt.ylabel('QTY')
plt.legend(['AMD', 'Intel'])
plt.xticks(number_ticks, df5.index, rotation=45);
```

3.2 Постройте график в относительных величинах

```python
df6 = df[df['Cpu']!='Samsung']
df6 = pd.crosstab(index=df['Company'],columns=df['Cpu'], normalize = 'index')
df6 = df6.drop(columns=['Samsung']).reset_index()

df6
```

```python
plt.figure(figsize=(6,4))

plt.bar(df6['Company'], df6['Intel'])
plt.bar(df6['Company'],df6['Intel'], bottom=df6['AMD'])
plt.title('Взаимосвязь')
plt.xlabel('Company')
plt.ylabel('Share')
plt.legend(['AMD', 'Intel'])
plt.xticks(rotation=45)
```
